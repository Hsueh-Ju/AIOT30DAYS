# 作業1
* 建立 DB 及 Collection (名稱自訂)，使用 Python 取得連線。

```python
from pymongo import MongoClient
client = MongoClient(host='127.0.0.1',port=27017)
db = client.hw1
collection = db.member

data_list = [
    {'name':'Peter','phone':'09111111','email':'test1@gmail.com','age':10},
    {'name':'George','phone':'09222222','email':'test2@gmail.com','age':20},
    {'name':'Mary','phone':'09333333','email':'test3@gmail.com','age':30}
]
collection.insert_many(data_list)
result = collection.find().limit(3)
for x in result:
    print(x)
```

# 作業2
* 使用 Python 操作 Mongo 中的資料。

```python
collection.update_one({'name':'Peter'},{'$set':{'name':'Frank'}})
result = collection.find().limit(3)
for x in result:
    print(x)
```

```python
collection.delete_one({'name':'Mary'})
result = collection.find().limit(2)
for x in result:
    print(x)
```
